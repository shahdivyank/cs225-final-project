```python
import numpy as np
import matplotlib.pyplot as plt

quadtree_times = np.load("quadtree_times.npy")
quadtree_times_copy = np.load("quadtree_times.npy")
```

```python
FOLDERS = [2, 4, 6, 8, 10]

plt.figure(figsize=(8, 5))

plt.plot(FOLDERS, quadtree_times_copy, linestyle='-', label="Quadtree Indexing Time")

plt.xlabel("Dataset Size (Million)")
plt.xticks(FOLDERS)
plt.ylabel("Time (seconds)")
plt.title("Quadtree Creation Performance")

plt.legend()
plt.show()
```

```python
positive_times_2 = np.load("2_positive_words_times.npy")
negative_times_2 = np.load("2_negative_words_times.npy")
positive_negative_times_2 = np.load("2_positive_negative_times.npy")

positive_times_4 = np.load("4_positive_words_times.npy")
negative_times_4 = np.load("4_negative_words_times.npy")
positive_negative_times_4 = np.load("4_positive_negative_times.npy")

positive_times_6 = np.load("6_positive_words_times.npy")
negative_times_6 = np.load("6_negative_words_times.npy")
positive_negative_times_6 = np.load("6_positive_negative_times.npy")
```

```python
plt.figure(figsize=(8, 5))

plt.plot(range(1, 6), positive_times_2, linestyle='-', label = "2 Million")
plt.plot(range(1, 6), positive_times_4, linestyle='-', label = "4 Million")
plt.plot(range(1, 6), positive_times_6, linestyle='-', label = "6 Million")

plt.xlabel("Number of Positive Words")
plt.xticks(range(1, 6))
plt.ylabel("Time (seconds)")
plt.title("Postive Words Performance")

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 5))

plt.plot(range(1, 6), negative_times_2, linestyle='-', label = "2 Million")
plt.plot(range(1, 6), negative_times_4, linestyle='-', label = "4 Million")
plt.plot(range(1, 6), negative_times_6, linestyle='-', label = "6 Million")

plt.xlabel("Number of Negative Words")
plt.xticks(range(1, 6))
plt.ylabel("Time (seconds)")
plt.title("Negative Words Performance")

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 5))

plt.plot(range(1, 6), positive_negative_times_2, linestyle='-', label = "2 Million")
plt.plot(range(1, 6), positive_negative_times_4, linestyle='-', label = "4 Million")
plt.plot(range(1, 6), positive_negative_times_6, linestyle='-', label = "6 Million")

plt.xlabel("Number of Positive/Negative Words")
plt.xticks(range(1, 6))
plt.ylabel("Time (seconds)")
plt.title("Postive/Negative Words Performance")

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 5))

plt.plot(range(1, 6), positive_negative_times_2, linestyle='-', label = "Positive and Negative")
plt.plot(range(1, 6), positive_times_2, linestyle='-', label = "Positive Only")
plt.plot(range(1, 6), negative_times_2, linestyle='-', label = "Negative Only")

plt.xlabel("Number of Positive/Negative Words")
plt.xticks(range(1, 6))
plt.ylabel("Time (seconds)")
plt.title("Postive/Negative Words Performance")

plt.legend()
plt.show()
```
